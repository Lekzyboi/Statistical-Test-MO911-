```python
# Necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
# Read the csv file, and check its top 5 rows
data = pd.read_csv('diamonds.csv')
```

```python
data.head()
```

```python
# Return the Series having unique values
x = data['cut'].unique()

# Return the Series having frequency count of each unique value
y = data['cut'].value_counts(sort=False)

plt.bar(x, y)

# Labeling the axes
plt.title('Bar Chart of Gold Count Plot')
plt.xlabel('cut')
plt.ylabel('count')

# Dsiplay the plot
plt.show()
```

```python
# Create bins with step-size 3000
bins = np.arange(0, data['price'].max()+3000, 3000)
plt.hist(data = data, x = 'price', bins = bins)
# Labeling the axes
plt.title('Histogram of Price of Gold')
plt.xlabel('Price')
plt.ylabel('count')

plt.show()
```

```python

plt.scatter(data = data, x = 'price', y = 'carat')
# Labeling the axes
plt.title('Scatterplot of Price of Gold against Carat')
plt.xlabel('Price')
plt.ylabel('count')

plt.show()
```
